```python
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
x= list(marks.keys())
y= list(marks.values())
```

```python
plt.bar(x,y)
plt.title("student grades")
```

```python
labels=x
plt.pie(y, labels=x)
plt.title("student grades")
plt.legend (bbox_to_anchor=(1.05,1.0))
```

```python
barx= np.array([1,3,4,5,7])
bary= np.array([1,2,3,4,5])
x= ('apples','pears','peaches','oranges','starfruit')
y= np.array([1,2,3,4,5])
linex1= np.array([2,4,6,8])
linex2= np.array([1,3,5,7])
liney1= np.array([1,4,5,8])
liney2= np.array([3,4,5,9])
histdata=np.random.randint(0,100, size=100)
```

```python
plt.subplot(2,3,1)
plt.plot(linex1,linex2,liney1,liney2)
plt.title("multiple line graph")

plt.subplot(2,3,2)
plt.bar(barx,bary)
plt.title("bar chart")

plt.subplot(2,3,3)
plt.pie(y, labels=x)
plt.title("pie chart")

plt.subplot(2,3,4)
plt.plot(linex1,linex2,liney1,liney2)
plt.grid()
plt.title("gridded chart")


plt.subplot(2,3,5)
plt.scatter(linex1,liney1)
plt.title("scatter plot")


plt.subplot(2,3,6)
plt.hist(histdata)
plt.title("histogram")
```
